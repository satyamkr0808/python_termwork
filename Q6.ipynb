```python
import networkx as nx
G = nx.Graph()

# Edges with costs
edges = [
    ('A', 'B', 1),
    ('A', 'C', 3),
    ('B', 'D', 3),
    ('B', 'E', 6),
    ('C', 'D', 1),
    ('D', 'E', 1)
]

for u, v, cost in edges:
    G.add_edge(u, v, weight=cost)

#  HEURISTIC VALUES  h(n)
heuristic = {
    'A': 7,
    'B': 6,
    'C': 4,
    'D': 2,
    'E': 0
}

def h(node, goal):
    return heuristic[node]

#  MAIN PROGRAM
print("   A* SEARCH using NetworkX Library")
print("\n  Graph edges (with costs):")
for u, v, data in G.edges(data=True):
    print(f"    {u} - {v}  :  cost = {data['weight']}")

print("\n  Heuristic values h(n):")
for node, val in heuristic.items():
    print(f"    h({node}) = {val}")
# Accept start node (goal is fixed as E)
print("\n  Available nodes: A  B  C  D  E")
start = input("  Enter START node : ").strip().upper()
goal  = 'E'

if start not in G.nodes:
    print(" Invalid node!")

else:
    
    #  nx.astar_path → gives OPTIMAL PATH directly
    path = nx.astar_path(G, start, goal,
                         heuristic=h,
                         weight='weight')

   
    # nx.astar_path_length → gives TOTAL COST directly
    total_cost = nx.astar_path_length(G, start, goal,
                                      heuristic=h,
                                      weight='weight')

   
    # Node expansion order
    print(f"  (i)   Node expansion order: {' → '.join(path)}")
    print(f"  (ii)  Optimal path: {' → '.join(path)}")
    print(f"  (iii) Total path cost: {total_cost}")
```
