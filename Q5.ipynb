```python
from collections import deque     
def bfs(graph, start, goal):
    """
    BFS uses a QUEUE (First In First Out)
    Explores all neighbours level by level → gives SHORTEST path
    """

    visited = []                
    queue = deque()           
    parent = {start: None}     

    queue.append(start)            

    print("\n  BFS Steps:")
    print("  " + "-" * 40)

    while queue:                       
        node = queue.popleft()           

        if node in visited:              
            continue

        visited.append(node)
        print(f"  Visiting: {node}   |   Queue: {list(queue)}")

        if node == goal:                  
            break

        for neigh in graph[node]:        
            if neigh not in visited and neigh not in queue:
                queue.append(neigh)
                if neigh not in parent:
                    parent[neigh] = node  

    #Reconstruct shortest path using parent dictionary
    path = []
    current = goal
    while current is not None:
        path.append(current)
        current = parent.get(current)
    path.reverse()

    #If path doesn't start from 'start',goal was unreachable
    if path[0] != start:
        return visited, None

    return visited, path
```

```python
# Static graph
def static_graph():
    print("STATIC GRAPH")
    graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D', 'E', 'F'],
        'C': ['A', 'D', 'F'],
        'D': ['B', 'C', 'E', 'F'],
        'E': ['B', 'D', 'F'],
        'F': ['C', 'D', 'E', 'B']
    }

    print("\n  Graph:", graph)
        
    # Accept start and goal from user
    print("\n  Available nodes: A B C D E F")
    start = input("  Enter START node : ").strip().upper()
    goal  = input("  Enter GOAL  node : ").strip().upper()

    if start not in graph or goal not in graph:
        print("  Invalid node entered")
        return

    visited, path = bfs(graph, start, goal)
    # Display traversal order
    print(f"  Traversal order  : {' → '.join(visited)}")
    # Display shortest path
    if path:
        print(f" Shortest path    : {' → '.join(path)}")
        print(f"       Path length      : {len(path) - 1} steps")
    else:
        print(f"  No path found from '{start}' to '{goal}'")
```

```python
# Dynamic graph
def dynamic_graph():
    print(" DYNAMIC GRAPH ")
    d = {}
    n = int(input("\n  Enter number of nodes: "))

    for i in range(n):
        u = input(f"  Enter node {i+1}: ").strip().upper()
        d[u] = []
        h = int(input(f"  Enter number of neighbours of '{u}': "))
        for j in range(h):
            v = input(f"    Neighbour {j+1}: ").strip().upper()
            d[u].append(v)

    print("\n  Graph:", d)

    # Accept start and goal from user
    print(f"\n  Available nodes: {' '.join(d.keys())}")
    start = input("  Enter START node : ").strip().upper()
    goal  = input("  Enter GOAL  node : ").strip().upper()

    if start not in d or goal not in d:
        print("  Invalid node entered")
        return

    visited, path = bfs(d, start, goal)

    # Display traversal order
    print(f" Traversal order  : {' → '.join(visited)}")
    # Display shortest path
    if path:
        print(f" Shortest path    : {' → '.join(path)}")
        print(f" Path length      : {len(path) - 1} steps")
    else:
        print(f"  No path found from '{start}' to '{goal}'")
```

```python
# Main menu
print("\n BFS GRAPH TRAVERSAL ")


while True:
    print("\n  1 → Static  Graph (6 nodes, 10 edges)")
    print("  2 → Dynamic Graph (you define nodes & edges)")
    print("  3 → Exit")

    choice = input("\n  Your choice: ").strip()

    if   choice == '1':
        static_graph()
    elif choice == '2':
        dynamic_graph()
    elif choice == '3':
        print("\n  Exit! \n")
        break
    else:
        print("  Enter 1, 2 or 3 only.")
```
