```python
vis = set()
path = []                          # Stores the path taken
 
def dfs(visit, graph, node, goal):
    if node not in visit:
        print(node, end=" ")       
        visit.add(node)
        path.append(node)          # Record the node in path
 
        if node == goal:           # Stop when goal is reached
            return True
 
        for neigh in graph[node]:
            if dfs(visit, graph, neigh, goal):
                return True        # Goal found down this branch, stop
 
        path.pop()                 # Backtrack if this path failed
    return False
```

```python
def static_graph():
    print("STATIC GRAPH")
    graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D', 'E', 'F'],
        'C': ['A', 'D', 'F'],
        'D': ['B', 'C', 'E', 'F'],
        'E': ['B', 'D', 'F'],
        'F': ['C', 'D', 'E', 'B']
    }
    print("\n  Graph:", graph)
    print("\n  Available nodes: A B C D E F")
    start = input("  Enter START node : ").strip().upper()
    goal  = input("  Enter GOAL  node : ").strip().upper()
 
    if start not in graph or goal not in graph:
        print(" Invalid node entered!")
        return
 
    # Reset visited and path before each run
    global vis, path
    vis  = set()
    path = []
    print("\n  Nodes visited: ", end="")
    found = dfs(vis, graph, start, goal)
 
    print()   # newline after visited nodes
    if found:
        print(f"  Path : {' → '.join(path)}")
    else:
        print(f"  No path found from '{start}' to '{goal}'")
```

```python
def dynamic_graph():
    
    print(" DYNAMIC GRAPH (User Input)")
  
    d = {}
    n = int(input("\n  Enter number of nodes: "))
 
    for i in range(n):
        u = input(f"  Enter node {i+1}: ").strip().upper()
        d[u] = []
        h = int(input(f"  Enter number of neighbours of '{u}': "))
        for j in range(h):
            v = input(f"    Neighbour {j+1}: ").strip().upper()
            d[u].append(v)
    
    print("\n  Graph:", d)
    print(f"\n  Available nodes: {' '.join(d.keys())}")
    start = input("  Enter START node : ").strip().upper()
    goal  = input("  Enter GOAL  node : ").strip().upper()
 
    if start not in d or goal not in d:
        print("  Invalid node entered!")
        return
        
    global vis, path
    vis  = set()
    path = []
 
    
    print("\n  Nodes visited: ", end="")
    found = dfs(vis, d, start, goal)
 
    if found:
        print(f" Path : {' → '.join(path)}")
    else:
        print(f" No path found from '{start}' to '{goal}'")
 
```

```python
print("DFS GRAPH TRAVERSAL")
while True:
    print("\n  1 → Static  Graph (6 nodes, 10 edges)")
    print("  2 → Dynamic Graph (you define nodes & edges)")
    print("  3 → Exit")
 
    choice = input("\n  Your choice: ").strip()
 
    if   choice == '1':
        static_graph()
    elif choice == '2':
        dynamic_graph()
    elif choice == '3':
        print("\n  Exit! \n")
        break
    else:
        print(" Enter 1, 2 or 3 only.")
```
